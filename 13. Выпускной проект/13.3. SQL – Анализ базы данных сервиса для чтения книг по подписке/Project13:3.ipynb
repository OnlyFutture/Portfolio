## Исследовательский анализ данных

### Cколько книг вышло после 1 января 2000 года

```python
query = '''
            SELECT COUNT(book_id)
            FROM books
            WHERE publication_date > '2000-01-01'
        '''
con=engine.connect()
df = pd.io.sql.read_sql(sql=text(query), con = con)
df
```

**Вывод**

После 1 января 2000 года вышло 819 книг.

### Количество обзоров и средняя оценка для каждой книги

```python
query = '''
            SELECT b.title,
                    COUNT(DISTINCT e.review_id) AS reviews,
                    ROUND(AVG(a.rating), 2) AS avg_rating                   
            FROM books b
            LEFT JOIN reviews e ON b.book_id = e.book_id
            LEFT JOIN ratings a ON b.book_id = a.book_id
            GROUP BY b.title
            ORDER BY reviews DESC
        '''
con=engine.connect()
df = pd.io.sql.read_sql(sql=text(query), con = con)
df          
```

**Вывод**

Максимально количество обзоров набрала книга Memoirs of a Geisha - 8 шт, со средним рейтингом 4,14.

### Издательство, которое выпустило наибольшее число книг толще 50 страниц

Необходимо учитывайть только книги толще 50 страниц — так мы исключим из анализа брошюры.

```python
query = '''
            SELECT p.publisher,
                    COUNT(b.book_id) AS count_books
            FROM publishers p
            LEFT JOIN books b ON p.publisher_id = b.publisher_id
            WHERE b.num_pages > 50
            GROUP BY p.publisher
            ORDER BY count_books DESC
            LIMIT 1
        '''

con=engine.connect()
df = pd.io.sql.read_sql(sql=text(query), con = con)
df 
```

**Вывод**

Издательство, которое выпустило наибольшее число книг - Penguin Books.

### Автор с самой высокой средней оценкой книг

Необходимо учитывайть только книги с 50 и более оценками.

```python
query = '''
            SELECT a.author,
                    AVG(r.rating) AS avg_rating
            FROM authors a
            LEFT JOIN books b ON a.author_id = b.author_id
            LEFT JOIN ratings r ON b.book_id = r.book_id
            WHERE b.book_id IN (SELECT book_id
                                FROM ratings
                                GROUP BY book_id
                                HAVING COUNT(rating) >=50)
            GROUP BY a.author   
            ORDER BY avg_rating DESC
            LIMIT 1
                                
        '''

con=engine.connect()
df = pd.io.sql.read_sql(sql=text(query), con = con)
df 
```

**Вывод**

Автор J.K. Rowling/Mary GrandPré имеет самую высокую среднюю оценку - 4,29.

### Среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
query = '''
                SELECT FLOOR(AVG(g.reviews)) AS avg_reviews
                FROM (
                        SELECT username,
                                COUNT(DISTINCT review_id) AS reviews
                        FROM reviews
                        WHERE username IN (
                                        SELECT username
                                        FROM ratings
                                        GROUP BY username
                                        HAVING COUNT(rating) > 48
                                        ) 
                        GROUP BY username
                ) AS g
        '''

con=engine.connect()
df = pd.io.sql.read_sql(sql=text(query), con = con)
df 
```

**Вывод**

Среднее количество обзоров от пользователей - 24.
